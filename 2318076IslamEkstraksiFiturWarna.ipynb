# 3. Melakukan ekstraksi fitur pada 10 citra (tergantung fitur apa yg ingin anda lakukan bisa fitur warna atau fitur tekstur)

```python
import cv2
import os
import numpy as np
import pandas as pd

# Tentukan path folder tempat citra disimpan
folder_path = '/content/drive/MyDrive/Dataset Daun'

# Buat list kosong untuk menyimpan nilai fitur warna dari setiap citra
features = []
# Buat list kosong untuk menyimpan nama file citra
image_names = []

# Loop untuk membaca setiap citra di dalam folder
for filename in os.listdir(folder_path):
    # Baca citra
    img = cv2.imread(os.path.join(folder_path, filename))
    # Hitung histogram untuk setiap kanal warna
    hist_r = cv2.calcHist([img], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([img], [1], None, [256], [0, 256])
    hist_b = cv2.calcHist([img], [2], None, [256], [0, 256])
    # Gabungkan histogram untuk setiap kanal warna menjadi satu array fitur warna
    color_feature = np.concatenate((hist_r, hist_g, hist_b), axis=None)
    # Tambahkan fitur warna ke dalam list
    features.append(color_feature)
    # Tambahkan nama file citra ke dalam list
    image_names.append(filename)

# Konversi list fitur warna dan nama file citra menjadi numpy array
features = np.array(features)
image_names = np.array(image_names)

# Buat DataFrame dari numpy array fitur warna dan nama file citra
df = pd.DataFrame({'Image': image_names, 'R': features[:,0], 'G': features[:,1], 'B': features[:,2]})

# Tampilkan DataFrame di layar
print(df)
```

# 4. Melakukan penampungan nilai-nilai fitur ke dalam file .xlsx atau .csv

```python
import cv2
import os
import numpy as np
import pandas as pd

# Tentukan path folder tempat citra disimpan
folder_path = '/content/drive/MyDrive/Dataset Daun'

# Buat list kosong untuk menyimpan nilai fitur warna dari setiap citra
features = []
# Buat list kosong untuk menyimpan nama file citra
image_names = []

# Loop untuk membaca setiap citra di dalam folder
for filename in os.listdir(folder_path):
    # Baca citra
    img = cv2.imread(os.path.join(folder_path, filename))
    # Hitung histogram untuk setiap kanal warna
    hist_r = cv2.calcHist([img], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([img], [1], None, [256], [0, 256])
    hist_b = cv2.calcHist([img], [2], None, [256], [0, 256])
    # Gabungkan histogram untuk setiap kanal warna menjadi satu array fitur warna
    color_feature = np.concatenate((hist_r, hist_g, hist_b), axis=None)
    # Tambahkan fitur warna ke dalam list
    features.append(color_feature)
    # Tambahkan nama file citra ke dalam list
    image_names.append(filename)

# Konversi list fitur warna dan nama file citra menjadi numpy array
features = np.array(features)
image_names = np.array(image_names)

# Buat DataFrame dari numpy array fitur warna dan nama file citra
df = pd.DataFrame({'Image': image_names, 'R': features[:,0], 'G': features[:,1], 'B': features[:,2]})

# Simpan DataFrame ke dalam file CSV
df.to_csv('fitur_warna.csv', index=False)
```

# 5. Nyoba Klasifikasi

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB

# Membaca data
data = pd.read_csv("/content/drive/MyDrive/Dataset Daun/fitur_warna.csv")

# Menentukan batas ambang untuk warna biru dan merah
threshold = 1200

# Mengelompokkan kupu-kupu berdasarkan warna biru dan merah
X = data[["R", "G", "B"]]
y = ["biru" if sum(rgb) < threshold else "merah" for rgb in X.values]

# Membuat model pengklasifikasi Naive Bayes
clf = GaussianNB()
clf.fit(X, y)

# Melakukan prediksi pada data
pred = clf.predict(X)
data["Warna"] = pred

# Menampilkan data yang telah dikelompokkan
print(data)
```
